#### Plot funciton x^3

```python
import numpy as np 
x = np.array([1,2,3,4,5,6])
y = x ** 3
plt.plot(x, y)
plt.show()
```

- We can use np.arange function which arranges the value of x so to get more smooth curve
- x = np.arange(starting, ending , increment)
- x = np.(0,5,0.1)
- x = [0,0.1,0.2,0.3.........4.9]

```python
x = np.arange(0,7,0.1)
y = x ** 3
plt.plot(x, y)
plt.show()
```

```python
a = [3,4,5]        #this will treated as y and y has 3 valuse so x will be taken as [0,1,2]
plt.plot(a)         
plt.show()
```

```python
x = np.arange(0,7,0.1)
y = x ** 3
plt.plot(x, y, color= "red" , #marker="o" ,
        linewidth=3.5)
plt.show()
```

```python
x = np.arange(0,7,0.1)
y = x ** 3
plt.plot(x, y , color="blue" , linewidth=3)
plt.ylabel("Cubes (X^3)")
plt.xlabel("Numbers")
plt.title("Graph of x^3")
plt.show()
```

```python
x = np.arange(0,7,0.1)
y = x ** 3
y1 = x ** 2
plt.plot(x, y , color="blue" , linewidth=3)
plt.plot(x, y1 , color="red" , linewidth=3)
plt.ylabel("Cubes (X^3) and Squares (x^2)")
plt.xlabel("Numbers")
plt.title("Graph of x^3")
plt.show()
```

- Apply legend

```python
x = np.arange(0,11,0.2)
y = x ** 3
y1 = x ** 2

plt.plot(x, y, color='blue' , linewidth = 3 , label = 'x^3')
plt.plot(x, y1, color='red', linewidth=3, label='x^2')
plt.title('Graphs of Cubes and Squares')
plt.ylabel('Y ->')
plt.xlabel('x ->')
plt.legend()
plt.show()
```

- Change the default axes

```python
x = np.arange(0,5,0.2)
y = x ** 3
y1 = x ** 2

plt.plot(x, y, color='blue', linewidth=4, label='x^3')
plt.plot(x, y1, color='red', linewidth=4, label='x^2')
plt.axis=[0,10,0,150]   # x axis and y axis
plt.title('Graphs for Cubes and Squares')
plt.ylabel('Y')
plt.xlabel('X')
plt.grid()   #create a grid
plt.legend()
plt.show()
```

- Add some text in the plot co-ordinates (x,y)

```python
x = np.arange(0,5,0.2)
y = x ** 3
y1 = x ** 2

plt.plot(x, y, color='blue', linewidth=4, label='x^3')
plt.plot(x, y1, color='red', linewidth=4, label='x^2')
plt.axis=[0,10,0,150]   # x axis and y axis
plt.title('Graphs for Cubes and Squares')
plt.ylabel('Y')
plt.xlabel('X')
plt.grid()   #create a grid

plt.text(4,64, 'Cube of 4' , fontsize=12)
plt.legend()
plt.show()
```

#### Pie Graph

```python
sizes = [3,4,6,2]         
plt.pie(sizes)
plt.show()
```

```python
sizes = [3,4,6,2]    
colors = ['blue','green' , 'red' , 'pink']     
labels = ['A','B','C','D']
plt.pie(sizes , colors=colors, labels=labels, autopct='%.2f%%')     #autoptc shows percentage upto 2 significant figures we uses %% to add percentage sign in the end
plt.title("Pie Chart Demo")
plt.show()
```

- Highlight a part in the pie graph
- explode is used

```python
sizes = [3,4,6,2]    
colors = ['blue','green' , 'red' , 'pink']     
labels = ['A','B','C','D']
explode = [0.1,0,0,0]           # this points describe the distance how far the part is from the rest
plt.pie(sizes , colors=colors, labels=labels, autopct='%.2f%%', explode=explode)     #autoptc shows percentage upto 2 significant figures we uses %% to add percentage sign in the end
plt.title("Pie Chart Demo")
plt.show()

#the explode function works in the counterclock wise directior to make is clockwise we need to set counterclock=false
```

```python
sizes = [3,4,6,2]    
colors = ['blue','green' , 'red' , 'pink']     
labels = ['A','B','C','D']
explode = [0.1,0,0,0] 
plt.pie(sizes , colors=colors, labels=labels, autopct='%.2f%%' , explode=explode, counterclock=False )     #autoptc shows percentage upto 2 significant figures we uses %% to add percentage sign in the end
plt.title("Pie Chart Demo")
plt.show()
```

- By make it clockwise the 0 deg is above a and the graph doesnot look proper 
- We can set the angle of start at any degree by using startangel=

```python
sizes = [3,4,6,2]    
colors = ['blue','green' , 'red' , 'pink']     
labels = ['A','B','C','D']
explode = [0.1,0,0,0] 
plt.pie(sizes , colors=colors, labels=labels, autopct='%.2f%%' , explode=explode, counterclock=False, startangle=45 )     #autoptc shows percentage upto 2 significant figures we uses %% to add percentage sign in the end
plt.title("Pie Chart Demo")
plt.show()
```
