### We can apply a function to a dataset to make changes of complex type

- We can use apply(function)
- apply() takes a function as an argument and return value and stores in a new column 

```python
def sex_determine(str):
    if str == 'Male' :
        return 0
    else :
        return 1
    
df['sex'] = df['Gender'].apply(sex_determine)
df.head()
```
